### walmartデータからAutoAI時系列予測のためのトレーニングデータに変換します

まず、画面右上のドットが縦に３つあるアイコンをクリックし、「プロジェクト・トークンの挿入」を選択して、追加されたセルを実行してください

```python
import pandas as pd
```

```python
# features.csvのロード
features_csv = project.get_file("features.csv")
features = pd.read_csv(features_csv)
features = features[features.Store == 1]
features
```

```python
# train.csvのロード
train_csv = project.get_file("train.csv")
train = pd.read_csv(train_csv)
train = train[train.Store == 1]
train
```

```python
# trainデータフレームをピボットしてDept(部門番号)ごとの週売上データに変換する
train_pivot_test = train.pivot(index="Date",columns="Dept",values="Weekly_Sales").add_prefix("週売上_")
train_pivot_test
```

```python
# train_pivotデータフレームにfeatureデータフレームを結合する
train_merged = features.merge(train_pivot_test, how='inner', on = ['Date']).reset_index(drop=True)
train_merged
```

```python
# 最後にプロジェクト内にデータ資産として保管する
project.save_data("train_merged.csv", train_merged.to_csv(header=True,index=False,na_rep='NA'), overwrite=True)
```
